```python
import datetime

# Task class to represent a project task
class Task:
    def __init__(self, title, description, start_date, end_date):
        self.title = title
        self.description = description
        self.start_date = start_date
        self.end_date = end_date
        self.completed = False

    def complete_task(self):
        self.completed = True

    def __str__(self):
        status = "Completed" if self.completed else "In Progress"
        return f"Task: {self.title}\nDescription: {self.description}\nStart Date: {self.start_date}\nEnd Date: {self.end_date}\nStatus: {status}"

# ProjectManagement class to manage tasks
class ProjectManagement:
    def __init__(self):
        self.tasks = []

    def add_task(self, task):
        self.tasks.append(task)

    def view_tasks(self):
        if not self.tasks:
            print("No tasks available.")
        else:
            for task in self.tasks:
                print(task)
                print("-" * 20)

    def mark_task_complete(self, task_title):
        for task in self.tasks:
            if task.title == task_title:
                task.complete_task()
                print(f"Task '{task_title}' marked as complete.")
                return
        print(f"Task '{task_title}' not found.")

# Create a project management instance
project_manager = ProjectManagement()

# Add tasks to the project
task1 = Task("Design Phase", "Complete the design of the project.", "2025-02-01", "2025-02-15")
task2 = Task("Development Phase", "Develop the core functionality.", "2025-02-16", "2025-03-01")

project_manager.add_task(task1)
project_manager.add_task(task2)

# View tasks
project_manager.view_tasks()

# Mark a task as complete
project_manager.mark_task_complete("Design Phase")

# View tasks again to see the updated status
project_manager.view_tasks()
```

```python
import tkinter as tk
from tkinter import messagebox
from tkinter import simpledialog

class Task:
    def __init__(self, title, description, start_date, end_date):
        self.title = title
        self.description = description
        self.start_date = start_date
        self.end_date = end_date
        self.completed = False

    def complete_task(self):
        self.completed = True

    def __str__(self):
        status = "Completed" if self.completed else "In Progress"
        return f"{self.title}\nDescription: {self.description}\nStart Date: {self.start_date}\nEnd Date: {self.end_date}\nStatus: {status}"

class ProjectManagement:
    def __init__(self):
        self.tasks = []

    def add_task(self, task):
        self.tasks.append(task)

    def view_tasks(self):
        return self.tasks

    def mark_task_complete(self, task_title):
        for task in self.tasks:
            if task.title == task_title:
                task.complete_task()
                return True
        return False

def add_task():
    title = simpledialog.askstring("Task Title", "Enter task title:")
    description = simpledialog.askstring("Task Description", "Enter task description:")
    start_date = simpledialog.askstring("Start Date", "Enter start date (YYYY-MM-DD):")
    end_date = simpledialog.askstring("End Date", "Enter end date (YYYY-MM-DD):")

    if title and description and start_date and end_date:
        task = Task(title, description, start_date, end_date)
        project_manager.add_task(task)
        messagebox.showinfo("Success", "Task added successfully!")
    else:
        messagebox.showerror("Error", "All fields are required!")

def view_tasks():
    tasks = project_manager.view_tasks()
    if not tasks:
        messagebox.showinfo("Tasks", "No tasks available.")
    else:
        task_list = "\n\n".join(str(task) for task in tasks)
        messagebox.showinfo("Tasks", task_list)

def complete_task():
    task_title = simpledialog.askstring("Complete Task", "Enter task title to mark as complete:")
    if project_manager.mark_task_complete(task_title):
        messagebox.showinfo("Success", f"Task '{task_title}' marked as complete.")
    else:
        messagebox.showerror("Error", f"Task '{task_title}' not found.")

# Create the main application window
app = tk.Tk()
app.title("Project Management Tool")

project_manager = ProjectManagement()

# Create buttons for the interface
add_task_button = tk.Button(app, text="Add Task", command=add_task)
view_tasks_button = tk.Button(app, text="View Tasks", command=view_tasks)
complete_task_button = tk.Button(app, text="Complete Task", command=complete_task)

# Arrange buttons in the window
add_task_button.pack(pady=10)
view_tasks_button.pack(pady=10)
complete_task_button.pack(pady=10)

# Run the application
app.mainloop()
```
