```python
import sqlite3
# Conecte-se ao banco de dados (ou crie um novo arquivo de banco de dados)
conexao = sqlite3.connect('dbPousada.db')
#Gerar nome da tabula SQLite
tabname = str('dbtb_' + arqname)
# Exportar o DataFrame para uma tabela no banco de dados SQLite
df.to_sql(tabname, conexao, if_exists='append', index=False)
# Fechar a conexão
conexao.close()
```

```python
# Conectar ao banco de dados SQLite
conn = sqlite3.connect('dbPousada.db')
sqltabname = str('SELECT * FROM ' +'"'+tabname+'"')

print (sqltabname)
# Ler a tabela para um DataFrame do pandas
tabsql = pd.read_sql(sqltabname, conn)
# Fechar a conexão
conn.close()
#apresenta alongamento max de dados e columnas
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)

# Mostrar o DataFrame
display (tabsql)
```
