```python
import pandas as pd
```

```python
df = pd.read_excel('C:/Projetos/Mercados/Trading_System/Database/ExperimentosParametros/ResumenExperimentos1H(03-10-22) - Periodo1.xlsx')
display (df.head())
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.express as px
import pandas as pd

#long2 = df.query('longindex >= 2')
long2 = df[df['long index'] > 2 ]
display (long2.head())

# Create a bar chart
bar_chart = px.bar(df, x=df['long index']  , y=  df['k'] , title='Bar Chart Example')

# Create a line chart
line_chart = px.line(df, x=df['long index']  , y=  df['k'], title='Line Chart Example')

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the app layout
app.layout = html.Div(children=[
    html.H1(children='Basic Dashboard'),

    html.Div(children='''
        Dash: A web application framework for Python.
    '''),

    dcc.Graph(
        id='bar-chart',
        figure=bar_chart
    ),

    dcc.Graph(
        id='line-chart',
        figure=line_chart
    )
])

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
