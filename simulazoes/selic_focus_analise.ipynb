```python
df = pd.read_excel("PrevSelic010216_300325.xls", header=1)
display(df.head())
```

```python
numcol = len(df.columns)
print (numcol)
print (df.columns[1])
def substituir_nome(coluna):
    # Obtém o ano após a barra
    if coluna[0] =='7' :
        ano = coluna.split('/')[-1]
        newcol = f"6/11/{ano}"
        return newcol  # Nova data com o ano
    elif coluna[0] =='8':
        ano = coluna.split('/')[-1]
        newcol = f"11/12/{ano}"
        return newcol
    elif coluna[0] =='1':
        ano = coluna.split('/')[-1]
        newcol = f"30/01/{ano}"
        return newcol
    elif coluna[0] =='2':
        ano = coluna.split('/')[-1]
        newcol = f"20/03/{ano}"
        return newcol
    elif coluna[0] =='3':
        ano = coluna.split('/')[-1]
        newcol = f"08/05/{ano}"
        return newcol
    elif coluna[0] =='4':
        ano = coluna.split('/')[-1]
        newcol = f"19/06/{ano}"
        return newcol
    elif coluna[0] =='5':
        ano = coluna.split('/')[-1]
        newcol = f"31/07/{ano}"
        return newcol
    elif coluna[0] =='6':
        ano = coluna.split('/')[-1]
        newcol = f"18/09/{ano}"
        return newcol
    else:
        return coluna  # Caso não siga o padrão, mantém o nome original

# Renomeando as colunas do DataFrame
df.rename(columns=substituir_nome, inplace=True)
display (df)



  
```

```python
import matplotlib.pyplot as plt

# Converte a primeira coluna para o tipo datetime
df['Data'] = pd.to_datetime(df['Data'])

# Ordena o DataFrame pela primeira coluna
df = df.sort_values(by='Data')

# Obtém os cabeçalhos para o eixo X (excluindo a primeira coluna)
eixo_x = df.columns[1:]

# Obtém os valores do último registro (exceto a primeira coluna)
last_prev = df.iloc[-1, 1:]
sixm_prev = df.iloc[-182, 1:]
year_prev = df.iloc[-365, 1:]
twoyear_prev = df.iloc[-730, 1:]
fouryear_prev = df.iloc[-1460, 1:]

# Criação do gráfico
plt.figure(figsize=(8, 5))

plt.plot(eixo_x, last_prev,label ='Last', color='black')
plt.plot(eixo_x, sixm_prev,label = 'Six Month', color='skyblue')
plt.plot(eixo_x, year_prev,label = 'One Year', color='red')
#plt.plot(eixo_x, twoyear_prev,label = 'Two Year', color='green')
#plt.plot(eixo_x, fouryear_prev,label = 'Four Year', color='yellow')

plt.xlabel('Data Previção')
plt.tight_layout()

plt.ylabel('Previções')
plt.title('Previções Focus Tasa Selic')
plt.grid(True)

# Rotação dos rótulos do eixo X para melhor legibilidade
plt.xticks(rotation=45)
plt.tight_layout()
plt.subplots_adjust(bottom=0.2)
plt.xticks(fontsize=8)
#plt.subplots(constrained_layout=True)

plt.legend()


plt.show()
```
