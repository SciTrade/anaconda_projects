```python
import pyodbc
import pandas as pd

# Definir o caminho para o banco de dados Access
caminho_bd = r'C:\Users\scitr\anaconda_projects\access_pandas_sqlite\Testeo_Multivariado.accdb'


# Definir a string de conex√£o
conn_str = (
    r'DRIVER={Microsoft Access Driver (*.mdb, *.accdb)};'
    r'DBQ=' + caminho_bd + ';'
)
# Conectar ao banco de dados
conn = pyodbc.connect(conn_str)
# Obter o nome de todas as tabelas
cursor = conn.cursor()
cursor.tables()
tabelas = [row.table_name for row in cursor if row.table_type == 'TABLE']

# Mostrar as tabelas
display (tabelas)
# Ler uma tabela para um DataFrame do pandas
df = pd.read_sql('SELECT * FROM Experimentos', conn)
display (df.head())
display (df.describe())
# Fechar a conex√£o
conn.close()
```
